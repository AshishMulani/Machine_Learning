```python
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori, association_rules
import ipywidgets as widgets
from ipywidgets import interact
from tqdm import tqdm 
import warnings
warnings.filterwarnings("ignore")
```

```python
fp_df = pd.read_csv("Datasets/Faceplate.csv", index_col=0)
fp_df
```

```python
fp_df = fp_df.astype(bool)
itemsets = apriori(fp_df, min_support=0.2, use_colnames=True)

rules = association_rules(itemsets, metric='confidence', min_threshold=0.6)
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

relv_rules = rules[rules['lift']>1]
relv_rules.sort_values('lift', ascending = False)
```

# Cosmetics Dataset

```python
cosm = pd.read_csv("Datasets/Cosmetics.csv", index_col=0)
cosm
```

```python
itemsets = apriori(cosm, min_support=0.2, use_colnames=True)

rules = association_rules(itemsets, metric='confidence', min_threshold=0.6)
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

relv_rules = rules[rules['lift']>1]
relv_rules.sort_values('lift', ascending = False)
```

```python
def gen_rules(permissible_support, confidence_threshold):
    itemsets = apriori(cosm, min_support=0.2, use_colnames=True)

    rules = association_rules(itemsets, metric='confidence', min_threshold=0.6)
    rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

    relv_rules = rules[rules['lift']>1]
    return relv_rules.sort_values('lift', ascending = False)
```

```python
interact(gen_rules, 
        permissible_support = widgets.FloatSlider(value=0.3, min=0.01, max=1, step=0.01,
                                                 description="Minimum Support:"),
         confidence_threshold = widgets.FloatSlider(value=0.3, min=0.01, max=1, step=0.01,
                                                 description="Confidence Threshold:"))
```
