```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import widget,interact
from sklearn.metrics import root_mean_squared_error
```

```python
df = pd.read_csv("D:/Machine_Learning/Datasets/monthly-milk-production-pounds-p.csv", index_col=0)
y_train = df['Milk'].iloc[:-12]
y_test = df['Milk'].iloc[-12:]
y_train.shape, y_test.shape
```

```python
plt.plot(y_train, label='Train', c="blue")
plt.plot(y_test, label='test', c="orange")
plt.legend(loc='best', frameon=True)
plt.show()
```

```python
def MA(span):
    fcast = y_train.rolling(span).mean()
    MA = fcast.iloc[-1]
    y_pred = pd.Series(MA.repeat)
    rmse = np.round(root_mean_squared_error(y_test, y_pred), 2)
    plt.text(164,900, f"RMSE = {rmse}")
    # plt.plot( df['Milk'], label='Data')
    # plt.plot(fcast, label = 'Centered Rolling Mean')
    plt.legend()
    plt.show()
interact(MA, span = (1,12)) 
```
