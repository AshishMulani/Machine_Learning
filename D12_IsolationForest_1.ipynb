## Removing Outliers

```python
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
# Assuming glass, predictions, LabelEncoder, and np are already defined/imported
in_glass = glass[predictions != -1]
le = LabelEncoder()
in_glass['Type'] = le.fit_transform(in_glass['Type'])
X, y = in_glass.drop('Type', axis=1), in_glass['Type']

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=26)

params = {
    'learning_rate': np.linspace(0.0001, 0.9, 30), 
    'n_estimators': [10, 25, 50, 75, 100],
    'max_depth': [2, 3, 4, 5, 6]
}

gbm = XGBClassifier(random_state=26, n_jobs=-1)
gcv = GridSearchCV(gbm, param_grid=params, cv=kfold, scoring='neg_log_loss')

gcv.fit(X, y)
gcv.best_params_, gcv.best_score_
```
