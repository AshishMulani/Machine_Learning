```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.compose import  ColumnTransformer, make_column_transformer, make_column_selector
from sklearn.metrics import r2_score, log_loss, accuracy_score, roc_auc_score, f1_score, log_loss

from sklearn.tree import DecisionTreeRegressor, plot_tree, DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression, ElasticNet
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier, StackingClassifier, StackingRegressor

from tqdm import tqdm  # Provides the progress of model running
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
glass = pd.read_csv('D:/Machine_Learning/Cases/Glass_Identification/Glass.csv')
glass
```

```python
# le = LabelEncoder()
# glass['Type'] = le.fit_transform(glass['Type'])
X , y = glass.drop(['Type'], axis=1), glass['Type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=26, stratify= y)
```

```python
dtc = DecisionTreeClassifier(random_state=26)
nb = GaussianNB()
lr = LogisticRegression()
rf = RandomForestClassifier(random_state=26)
svm = SVC(random_state=26)

stack = StackingClassifier(estimators= [('Tree', dtc),('NB', nb), ('LR', lr)], final_estimator= rf)  

stack.fit(X_train, y_train)
y_pred = stack.predict(X_test)
f1_score(y_test, y_pred, average='macro')
```

```python
stack = StackingClassifier(estimators= [('Tree', dtc),('NB', nb), ('LR', lr)], final_estimator= rf, passthrough=True)   # passthrough=False by default
stack.fit(X_train, y_train)
y_pred = stack.predict(X_test)
f1_score(y_test, y_pred, average='macro')
```

```python
xgbm = XGBClassifier(random_state=26)
stack = StackingClassifier(estimators= [('Tree', dtc),('RF', rf),('NB', nb), ('SVM', svm)], final_estimator= xgbm, passthrough=True)
stack.fit(X_train, y_train)
y_pred = stack.predict(X_test)
f1_score(y_test, y_pred, average='macro')
```
