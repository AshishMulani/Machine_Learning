```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, KFold, cross_val_score  #***
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.compose import  ColumnTransformer, make_column_transformer, make_column_selector
from sklearn.metrics import r2_score, log_loss, accuracy_score, roc_auc_score, f1_score, log_loss

from sklearn.tree import DecisionTreeRegressor, plot_tree, DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression, ElasticNet, Ridge
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from xgboost import XGBClassifier, XGBRegressor
from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, StackingClassifier, StackingRegressor

from tqdm import tqdm  # Provides the progress of model running
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
concrete = pd.read_csv('D:/Machine_Learning/Cases/Concrete_Strength/Concrete_Data.csv')
X , y = concrete.drop(['Strength'], axis=1), concrete['Strength']

lr = LinearRegression()
kfold = KFold(n_splits=5, shuffle=True, random_state=26)
results = cross_val_score(lr, X, y, cv=kfold)  # Finds r2_score by default
print(results)
print(np.mean(results))
```

```python

alphas = np.linspace(0.001, 10, 30)
scores=[]
for a in tqdm(alphas):
    ridge = Ridge(alpha=a)
    kfold = KFold(n_splits=5, shuffle=True, random_state=26) 
    scores.append([a, cross_val_score(ridge, X, y, cv=kfold).mean()])
    
df_scores = pd.DataFrame(scores, columns=['alpha', 'cross_val_score' ])
df_scores.sort_values(['cross_val_score'], ascending=False).head()

```

```python
alphas = np.linspace(0.001, 10, 30)
ratios = np.linspace(0.001,1,10) 
scores=[]
for a in tqdm(alphas):
    for r in ratios:
        el = ElasticNet(alpha=a, l1_ratio=r)
        results = cross_val_score(el, X, y, cv=kfold) 
        scores.append([a, r, np.mean(results)])
        print(results)
    
```

```python
df_scores = pd.DataFrame(scores, columns=['alpha', 'l1_ratio', 'cross_val_score' ])
df_scores.sort_values(['cross_val_score'], ascending=False).head()
```

### changing scoring parameter
#### `scoring` as RMSE

```python
# from string_name_scores
alphas = np.linspace(0.001, 10, 30)
ratios = np.linspace(0.001,1,10) 
scores=[]
for a in tqdm(alphas):
    for r in ratios:
        el = ElasticNet(alpha=a, l1_ratio=r)
        results = cross_val_score(el, X, y, cv=kfold, scoring='neg_root_mean_squared_error')  # why 'neg': optimizer uses bigger the better in negative way
        scores.append([a, r, np.mean(results)])
df_scores = pd.DataFrame(scores, columns=['alpha', 'l1_ratio', 'root_mean_squared_error' ])
df_scores.sort_values(['root_mean_squared_error'], ascending=False).head()       
```

```python
features = [2,3,4,5,6,7]
scores=[]

for f in features:
    rf = RandomForestRegressor(random_state=26, max_features=f )
    results = cross_val_score(rf, X, y, cv=kfold, scoring='neg_root_mean_squared_error') 
    scores.append([f, np.mean(results)])
df_scores = pd.DataFrame(scores, columns=['features', 'root_mean_squared_error' ])
df_scores.sort_values(['root_mean_squared_error'], ascending=False).head()
```

```python
housing = pd.read_csv('D:/Machine_Learning/Datasets/Housing.csv')
X , y = housing.drop(['price'], axis=1), housing['price']
```

```python
# to increase the speed remove sparse_output, set_output

# ohe=OneHotEncoder(sparse_output=False,drop="first").set_output(transform="pandas")
# trans = ColumnTransformer(transformers=[("OHE", ohe, make_column_selector(dtype_include=object))],     
#                           remainder="passthrough",verbose_feature_names_out=False).set_output(transform="pandas")

ohe=OneHotEncoder(drop="first")

trans = ColumnTransformer(transformers=[("OHE", ohe, make_column_selector(dtype_include=object))],     
                          remainder="passthrough", verbose_feature_names_out=False)

```

```python
features = [2,3,4,5,6,7]
scores=[]

for f in tqdm(features):
    rf = RandomForestRegressor(random_state=26, max_features=f )
    pipe = Pipeline([('OHE',trans), ('RF',rf)])
    results = cross_val_score(pipe, X, y, cv=kfold, scoring='neg_root_mean_squared_error') 
    scores.append([f, np.mean(results)])
df_scores = pd.DataFrame(scores, columns=['features', 'root_mean_squared_error' ])
df_scores.sort_values(['root_mean_squared_error'], ascending=False).head()
```
