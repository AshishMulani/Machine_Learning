```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import classification_report, roc_auc_score, log_loss
from sklearn.metrics import f1_score, accuracy_score, recall_score
from tqdm import tqdm 

from sklearn.naive_bayes import GaussianNB

import os
os.chdir("D:/Machine_Learning/Cases") 

import warnings
warnings.filterwarnings("ignore")
```

```python
sonar = pd.read_csv("Sonar/Sonar.csv")
sonar.head()
```

```python
le = LabelEncoder()
sonar['Class'] = le.fit_transform(sonar['Class'])
X , y = sonar.drop('Class' , axis = 1) , sonar['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=26, stratify=y)
```

```python
nb = GaussianNB()
nb.fit(X_train,y_train) # Calculates all the apriori probablitites (Test)
y_pred_prob = nb.predict_proba(X_test) # Calculates posterior probabilites (Validation)
print("ROC AUC =", roc_auc_score(y_test, y_pred_prob[:,1]))
y_pred = nb.predict(X_test)
print("Accuracy Score=", accuracy_score(y_test, y_pred))
```

### Glass Identification Dataset

```python
glass = pd.read_csv('Glass_Identification/Glass.csv')
glass.head(5)
```

```python
le = LabelEncoder()
glass['Type'] = le.fit_transform(glass['Type'])
X , y = glass.drop('Type' , axis = 1) , glass['Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=26, stratify=y)
```

```python
nb = GaussianNB()
nb.fit(X_train,y_train) # Calculates all the apriori probablitites (Test)
y_pred_prob = nb.predict_proba(X_test) # Calculates posterior probabilites (Validation)
print("Log Loss =", log_loss(y_test, y_pred_prob)) # Because of "multiclass error" we can't use ROC_auc_accuracy.
y_pred = nb.predict(X_test)
print("Accuracy Score=", accuracy_score(y_test, y_pred))
```
