```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import classification_report, roc_auc_score, log_loss
from sklearn.metrics import f1_score, accuracy_score, recall_score
from tqdm import tqdm 

from sklearn.naive_bayes import BernoulliNB

import os
os.chdir("D:/Machine_Learning/Cases/cancer") 

import warnings
warnings.filterwarnings("ignore")
```

### Cancer

```python
cancer = pd.read_csv("Cancer.csv", index_col=0)
le = LabelEncoder()
cancer['Class']= le.fit_transform(cancer['Class'])

X, y = cancer.drop('Class', axis=1), cancer['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=26, stratify=y)
```

```python
ohe=OneHotEncoder(sparse_output=False,drop="first", handle_unknown='ignore').set_output(transform="pandas")
X_trn_ohe = ohe.fit_transform(X_train)
X_tst_ohe = ohe.transform(X_test)
```

```python
nb = BernoulliNB()
nb.fit(X_trn_ohe, y_train)     # Calculate all the APRIORI Probabilities
y_pred_prob = nb.predict_proba(X_tst_ohe)    # Calculates POSTERIOR Probabilities-- acctualy finds baiyes prob..
print("ROC AUC =", roc_auc_score(y_test, y_pred_prob[:,1]))
y_pred = nb.predict(X_tst_ohe)
print("Accuracy Score=", accuracy_score(y_test, y_pred))
```

```python
log_loss(y_test, y_pred_prob)
```
